```python
location_columns = ['neighbourhood','city','latitude','longitude']
```

```python
unique_locations = df_Listings[location_columns].drop_duplicates().reset_index(drop=True)
```

```python
unique_locations['location_id'] = unique_locations.index + 1
```

```python
unique_locations
```

```python
df_Listings = df_Listings.merge(unique_locations, on=location_columns, how='left')
```

```python
df_Listings
```

```python
Listings_location=unique_locations[['location_id','neighbourhood','city','latitude','longitude']]
```

## Data convertion to SQL tables

```python
import sqlite3
```

```python
connect = sqlite3.connect('AirBnb Normalized Data.db')
```

```python
Host_information.to_sql('Host_information', connect, if_exists='append', index=False)
```

```python
Listings_location.to_sql('Listings_location',connect,if_exists='append',index=False)
```

```python
Listings.to_sql('Listings',connect, if_exists='append', index=False)
```

```python
Reviews.to_sql('Reviews', connect, if_exists='append', index=False)
```

```python
Listings_review_scores.to_sql('Listings_review_scores', connect, if_exists='append', index=False)
```

```python
connect.commit()
connect.close()
```
